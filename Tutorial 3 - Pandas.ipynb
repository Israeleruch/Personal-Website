## Modifying the Data

### The apply() Method

```python
#This method applies a given function to each column.

df[["Birth_Rate", "Internet_Users"]].apply(max)
```

```python
#We can also apply a function to each row by passing "axis = 1" as an argument.

df[["Birth_Rate", "Internet_Users"]].apply(max, axis = 1)
```

```python
#We can use the method to apply a lambda function.

df[["Birth_Rate", "Internet_Users"]].apply(lambda x: x*100)
```

```python
#Or we can apply a function we defined earlier.

def mean_division(col):
    return col / col.mean()

df[["Birth_Rate", "Internet_Users"]].apply(mean_division)
```

```python
def third_world(row):
    if row.Internet_Users < 30 and row.Birth_Rate > 40:
        row.Group = "Third World"
    return row

df.apply(third_world, axis = 1)
```

```python
#To save the modifications, simply overwrite the existing dataframe.

df = df.apply(third_world, axis = 1)
df
```

### The applymap() Method

```python
#This method applies a given function to each entry individually.

df.applymap(lambda k: np.sin(k) if type(k) != str else k)
```

```python
#Again, we can apply a function we defined earlier.

def change_entry(entry):
    if type(entry) != str:
        entry = entry**2 - 3*entry
    return entry

df[["Birth_Rate", "Internet_Users"]].applymap(change_entry)
```

### Inserting Missing Values

```python
#Let's first insert a few NA values.

random_rows = np.random.choice(np.arange(0, 195), size = 50, replace = False)
random_columns = np.random.choice(np.arange(0, 4), size = 50, replace = True)

for index in list(zip(random_rows, random_columns)):
    df.iloc[index] = np.nan

df
```

```python
#How many NA values does the dataframe contain?

df.isna().sum().sum()
```

```python
#One way to get rid of (some of) the NA values is to apply a function that replaces these values with something else.

def replace_nan(entry):
    if type(entry) != str and np.isnan(entry):
        return 0
    else:
        return entry
    
df = df.applymap(replace_nan)
```

## Grouping and Visualizing Data

```python
df = pd.read_csv("C:/Users/eruch/Dropbox/Macroeconomic Theory II/TA Section/2023/Python/Code Files/countries_data.csv")
```

```python
#We can group data based on a specific column or columns, enabling us to perform aggregate operations or apply functions on those groups.

group_count = df.groupby("Group")["Birth_Rate", "Internet_Users"].count()
group_count
```

```python
group_means = df.groupby("Group")["Birth_Rate", "Internet_Users"].mean()
group_means
```

```python
#Let's create a simple bar graph based on the means calculated above.

fig, ax = plt.subplots(figsize = (12, 9))

ax.bar(group_means.index - 0.2, group_means["Birth_Rate"], color = "b", width = 0.4, alpha = 0.7, label = "Birth Rate")
ax.bar(group_means.index + 0.2, group_means["Internet_Users"], color = "r", width = 0.4, alpha = 0.7, label = "Internet Users")

ax.set_xlabel("Group", font = "Serif", fontsize = 12)
ax.set_ylabel("Mean Values", font = "Serif", fontsize = 12)
ax.set_xticks(group_means.index)
ax.set_title("Mean Birth Rate and Internet Users by Group", font = "Serif", fontsize = 16)
ax.legend()

plt.show()
```

```python
#Now, let's create a simple scatter plot of the relationship between a country's birth rate and the share of internet users, for each income group.

fig, ax = plt.subplots(figsize = (12, 10))

colors = ["r", "b", "g", "black"]

for i in range(1, 5):
    income_group = df[df["Group"] == i]
    ax.scatter(income_group["Birth_Rate"], income_group["Internet_Users"], color = colors[i-1], label = "Income Group {}".format(i))
    
ax.set_xlabel("Birth Rate", font = "Serif", fontsize = 12)
ax.set_ylabel("Internet Users", font = "Serif", fontsize = 12)
ax.legend()

plt.show()
```

## Running Simple Regressions

```python
#Let's regress a country's birth rate on the share of internet users and a set of income group dummy variables.

#We import the necessary library to perform an OLS regression.
import statsmodels.api as sm

#We add three columns represneting the three highest income groups: lower middle, upper middle, and high.
for column, group in zip(["Lower_Middle", "Upper_Middle", "High"], [2, 3, 4]):
    df[column] = df["Group"].apply(lambda value: 1 if value == group else 0)
    
#We add three columns representing the interaction terms.
for interaction, column in zip(["IU_Lower_Middle", "IU_Upper_Middle", "IU_High"], ["Lower_Middle", "Upper_Middle", "High"]):
    df[interaction] = df[column] * df["Internet_Users"]
    
#We add a column for the intercept.
df["Intercept"] = 1

#We define the dependent variable and a vector containing the explanatory variables.
y, X = df["Birth_Rate"], df[["Intercept", "Internet_Users", "Lower_Middle", "Upper_Middle", "High", "IU_Lower_Middle", "IU_Upper_Middle", "IU_High"]]

#We fit the OLS model
model = sm.OLS(y, X)

#We make sure to calculate robust standard errors.
results = model.fit(cov_type = "HC3")

#Finally, we present the results.
results.summary()
```
