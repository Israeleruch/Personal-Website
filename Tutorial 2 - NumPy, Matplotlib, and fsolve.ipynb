### Practice Set 1

```python
#Define a function that generates an array of n normally-distributed elements and returns the mean of all positive elements.

def normal(n):
    array = np.random.randn(n)
    return array[array > 0].mean()
    
normal(100)
```

```python
#Define a function that multiplies two matrices and returns the positive eigenvalues of the product.

def eigen(a, b):
    product = a @ b
    e_v =  np.linalg.eigvals(product)
    return e_v[e_v > 0]

x = np.array([[2, -4], [1, 5]])
y = np.array([[0, 3], [4, 1]])

eigen(x, y)
```

```python
#Define a function that creates an array of n evenly spaced numbers in [0,10] and returns their cosine.

def cos_array(n):
    array = np.linspace(0, 10, n)
    return np.cos(array)

cos_array(10)
```

## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Let's create a simple plot.

import matplotlib.pyplot as plt #Importing the matplotlib package.
import numpy as np #Importing the NumPy package.

x = np.linspace(0, 10, 200) #Creating an evenly spaced NumPy array.
y = np.sin(x) #Taking the sine of each element.

fig, ax = plt.subplots() #Initializing a figure and axes objects.
ax.plot(x, y, "r") #Creating a plot containing three arguments: x-axis, y-axis, and color.
plt.show() #Presenting the plot.
```

```python
#Adding new features: the figure's size, the line's width, and a different color.

x = np.linspace(0, 10, 100) 
y = np.cos(x)

fig, ax = plt.subplots(figsize = (12, 8)) #Passing a tuple containing the figure's size as input.
ax.plot(x, y, "indigo", linewidth = 3) #Specifying a different color and the line's width.
plt.show()
```

```python
#Adding new features: label, alpha, and legend.

x = np.linspace(0, 3, 100) 
y = np.exp(x)

fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(x, y, "g", linewidth = 3, label = "Exp(x)", alpha = 0.5) #Adding a label and specifying the line's transperency.

ax.legend()
plt.show()
```

```python
#Adding new features: axes labels, ticks, title, and the legend's location.

x = np.linspace(0, 10, 300) 
y = (x**2 - 5*x + 2)/(x + 1)

fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(x, y, "chocolate", linewidth = 3, label = "Our Function", alpha = 0.8)

ax.set_xlabel("x") #Specifying the x-axis label.
ax.set_ylabel("y") #Specifying the y-axis label.
ax.set_xticks([0, 5, 10]) #Passing a list to determine the x-axis ticks.
ax.set_yticks(np.arange(-1, 5.5, 0.5)) #Passing a list to determine the y-axis ticks.
ax.legend(loc = "lower right") #Specifying the legend's location.
ax.set_title("Our Graph")
plt.show()
```

```python
#Multiple plots in a single figure.

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots(figsize = (12, 9))

for slope in np.linspace(1, 5, 12): #Creating a "for" loop with twelve different slopes.
    y = slope * x #The slope determines y.
    our_label = "m = {}".format(slope) #The label is the slope.
    ax.plot(x, y, label = our_label, alpha = 0.6, linewidth = 7 - slope)
    
ax.set_xlabel("x") 
ax.set_ylabel("y")
ax.set_xticks(np.arange(0, 11, 1)) 
ax.legend(loc = "upper left")
ax.set_title("Multiple Graphs")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(x, y, "sienna", linewidth = 3, label = "$y = sin(x)$", alpha = 0.7)
ax.plot(x, z, "black", linewidth = 3, label = "$z = cos(x)$", alpha = 0.7)

ax.legend(loc = "upper right")
ax.set_yticks([-1, 0, 1])
ax.set_xticks(np.linspace(0, 10, 5))
ax.set_title("Graphing Sine and Cosine Functions")
plt.show()
```

```python
#Plot two normally-distributed arrays with a mean of zero but different standard deviations.

x = np.random.normal(loc = 0, scale = 1, size = 100)
y = np.random.normal(loc = 0, scale = 3, size = 100)

fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(x, "black", alpha = 0.8, linewidth = 4)
ax.plot(y, "r", alpha = 0.5, linewidth = 2)

ax.set_xticks(np.arange(0, 125, 25)) 
ax.set_yticks(np.arange(-8, 10, 2)) 
ax.set_title("Normal Distributions", fontsize = 18, font = "Serif")
plt.show()
```

## SciPy - fsolve

```python
#First, import the desired packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#Define a function
def f(x):
    return x**3 - 2*x**2 - 3*x + 2

#Use fsolve to find a root
root = fsolve(f, 0)

#Present the root
root
```

```python
#Let's plot the function

x = np.linspace(-3, 4, 100)
y = f(x)
z = np.zeros(100)

fig, ax = plt.subplots(figsize = (12, 9))
ax.plot(x, y, "black", linewidth = 3, alpha = 0.7, label = "$f(x)$")
ax.plot(x, z, "olive", linewidth = 3, alpha = 0.7, label = "$y=0$")
        
ax.set_xticks(np.arange(-3, 4.5, 0.5))
ax.legend(loc = "upper left")
ax.set_xlabel("x", fontsize = 12, font = "Serif")
ax.set_ylabel("y", fontsize = 12, font = "Serif")
ax.set_title("Plotting $y = f(x)$", fontsize = 18, font = "Serif")

plt.show()
```

```python
#We see that there exist three roots, but we found only one. We can try to find all three by passing three starting estimates

roots = fsolve(f, [-1.5, 0.5, 3])
roots
```

```python
#Finally, we can check whether the function's value at the root is close to zero

np.isclose(f(roots), [0, 0, 0])
```

```python
#Now, let's solve a system of equations

def g(x):
    return [x[0]**2 - 3*x[1] + 2,
           -2*x[0] + x[1] + 1]

root = fsolve(g, [0, 0])
root
```

```python
#Again, plot the functions

x = np.linspace(-6, 6, 100)
y = [(x**2 + 2)/3, 2*x - 1]

fig, ax = plt.subplots(figsize = (12, 9))
ax.plot(x, y[0], "black", linewidth = 3, alpha = 0.7, label = "$y_1(x)$")
ax.plot(x, y[1], "olive", linewidth = 3, alpha = 0.7, label = "$y_2(x)$")
        
ax.set_xticks(np.arange(-6, 6.5, 1))
ax.legend(loc = "lower right")
ax.set_xlabel("x", fontsize = 12, font = "Serif")
ax.set_ylabel("y", fontsize = 12, font = "Serif")

plt.show()
```

```python
#As before, we see that there exist two roots. Let's try to find the other root.

other_root = fsolve(g, [5, 5])
other_root
```

```python
#The following equations charactarize a closed economy:
#(1) Y = C + I
#(2) C = 100 + 0.6Y
#(3) I = 200 - 10r
#(4) 600 = Y - 50r
#Using fsolve, find the equuilibrium output (Y), consumption (C), investment (I), and rent (r).

def economy(x):
    return [x[0] - x[1] - x[2],
           x[1] - 100 - 0.6*x[0],
           x[2] - 200 -10*x[3],
           600 - x[0] + 50*x[3]]

root = fsolve(economy, [1000, 500, 500, 1])
root
```
