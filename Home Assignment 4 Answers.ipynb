## Question 1

Answer Question 4.3.a from the Solow Assignment.

```python
import numpy as np
import matplotlib.pyplot as plt

#Parameter Values
delta = 0.15
alpha = 0.2
n = 0.05
s = 0.35
k_0 = 0.1

#Defining the series
k_array = np.array([k_0])
time_range = np.arange(0, 51, 1)
k = k_0

#Using a for loop to generate the sequence of k
for t in time_range[0:50]:
    k = ((1 - delta)*k + s*k**alpha)/(1 + n)
    k_array = np.append(k_array, k)
    
#Plotting the figure
fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(time_range, k_array, "r", linewidth = 3, label = "$k_t$", alpha = 0.8)

ax.set_yticks(np.arange(0, 2.1, 0.5))
ax.legend(loc = "upper left", fontsize = 12)
ax.set_title("The Evolution of the Capital-Labor Ratio", fontsize = 18, font = "Serif")
ax.set_xlabel("$t$", fontsize = 12)
ax.set_ylabel("$k_t$", fontsize = 12)
plt.show()
```

Answer Question 4.3.b from the Solow Assignment.

```python
#Defining the steady state equilibrium
k_bar = (s/(n+delta))**(1/(1-alpha))

#Checking whether our sequence reached the proximity of the steady state equilibrium
round(k_bar, 2) == round(k_array[-1], 2)
```

Answer Question 4.3.c from the Solow Assignment.

```python
#Defining the 45-degree line and the phi function
the_45_line = np.linspace(0, 5, 100)
phi_array = ((1 - delta)*the_45_line + s*the_45_line**alpha)/(1 + n)

#Plotting them in the same figure
fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(the_45_line, the_45_line, "r--", linewidth = 2, label = "$k_{t+1} = k_t$", alpha = 0.8)
ax.plot(the_45_line, phi_array, "b", linewidth = 4, label = "$\phi(k_t)$", alpha = 0.8)

ax.legend(loc = "upper left", fontsize = 12)
ax.set_title("The Dynamic System", fontsize = 18, font = "Serif")
ax.set_xlabel("$k_t$", fontsize = 12)
ax.set_ylabel("$k_{t+1}$", fontsize = 12)
plt.show()
```

## Question 2

The dynamic system is charactarized by the following dynamic equations:


\begin{equation*}
\begin{cases}
\dot{x} = 2x^2 - 3x + y\\
\dot{y} = -y^2 +5y -10x + 1
\end{cases}
\end{equation*}

Find the system's steady states.

```python
import numpy as np
from scipy.optimize import fsolve

def dynamic_system(x):
    return [2*x[0]**2 - 3*x[0] + x[1],
        -x[1]**2 + 5*x[1] - 10*x[0] + 1]

steady_state_1 = fsolve(dynamic_system, [-1, -1])
steady_state_2 = fsolve(dynamic_system, [1, 1])

steady_state_1, steady_state_2
```

Show that the values you obtained are indeed the system's steady states.

```python
np.isclose(dynamic_system(steady_state_1), [0, 0]), np.isclose(dynamic_system(steady_state_2), [0, 0])
```
